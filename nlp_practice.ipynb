```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize

# Download the NLTK sentence tokenizer
# nltk.download()
nltk.download('punkt')

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/India"

# Send a request to the webpage
response = requests.get(url)

# Parse the content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the main content of the article
content = soup.find_all('p')

# Combine all paragraphs into one string
paragraph = ""
for lines in content:
    paragraph += lines.text

# Tokenize the text into sentences
sentences = sent_tokenize(paragraph)
print('paragraph', paragraph)
print('sentences', sentences)

# Print the first 5 sentences as an example
for sentence in sentences[:5]:
    print(sentence)
```

```python
words = nltk.word_tokenize(paragraph)
print(words)
```

### Stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stemmer = PorterStemmer()
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
sentences_stem = sent_tokenize(paragraph)
for i in range(len(sentences_stem)):
    words = nltk.word_tokenize(sentences_stem[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences_stem[i] = ' '.join(words)
```

```python
print(sentences_stem)
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
sentences_lemma = sent_tokenize(paragraph)
for i in range(len(sentences_lemma)):
    words = nltk.word_tokenize(sentences_lemma[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences_lemma[i] = ' '.join(words)
```

```python
print(sentences_lemma)
```

### Bag of Words

```python
# Library for regular expression
import re
```

```python
sentences_bow = sent_tokenize(paragraph)
```

```python
corpus = []

for i in range(len(sentences_bow)):
    review = re.sub('[^a-zA-Z]', ' ', sentences_bow[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
print(x)
```

### Word2Vec

```python
from gensim.models import Word2Vec
```

```python
text = re.sub(r'\[[0-9]*\]', ' ', paragraph)
text = re.sub(r'\s+', ' ', text)
text = text.lower()
text = re.sub(r'\d', ' ', text)
text = re.sub(r'\s+', ' ', text)
```

```python
sentences_w2v = nltk.sent_tokenize(text)
sentences_w2v = [nltk.word_tokenize(sentence) for sentence in sentences_w2v]
```

```python
print(sentences_w2v)
```

```python
for i in range(len(sentences_w2v)):
    sentences_w2v[i] = [word for word in sentences_w2v[i] if word not in stopwords.words('english')]
```

```python
model = Word2Vec(sentences_w2v, min_count=1)
```

```python
# For version less than 4.0.0
# words = model.wv.vocab 
words = set(model.wv.key_to_index.keys())
```

```python
print(words)
```
